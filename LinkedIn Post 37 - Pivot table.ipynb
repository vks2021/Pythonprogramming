```python
import numpy as np
import pandas as pd
df = pd.read_csv("T20-ICC-WC.csv")
```

```python
df
```

```python
df1= df[['team_1', 'team_2', 'Winner_toss', 'Toss_descision', 'time', 'venue', 'target_achieved', 'Winner']]
```

```python
df1
```

```python
df1.pivot_table(index='Toss_descision', columns = ['venue', 'time'])
```

```python
df1.pivot_table(index='Toss_descision', columns = ['venue', 'time'], values = 'target_achieved')
```

```python
df1.pivot_table(index='Toss_descision', columns = ['venue', 'time'], values = 'Winner', aggfunc = 'count')
```

```python
df1.pivot_table(index='Toss_descision', columns = ['venue', 'time'], values = 'target_achieved', aggfunc = 'count')
```

```python
df1.pivot_table(index='Toss_descision', columns = ['venue', 'time'], values = 'target_achieved', aggfunc = 'count', margins = True)
```
