# Pivot Table 

```python
import numpy as np
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/vks2021/Pythonprogramming/main/T20-ICC-WC.csv')
df.head()
```

```python
df
```

```python
df1= df[['team_1', 'team_2', 'Winner_toss', 'Toss_descision', 'time', 'venue', 'target_achieved', 'Winner']]
```

```python
df1.head()
```

```python
df1.pivot_table(index='Toss_descision', columns = ['venue', 'time'])
```

```python
df1.pivot_table(index='Toss_descision', columns = ['venue', 'time'], values = 'target_achieved', fill_value=0)
```

```python
df1.pivot_table(index='Toss_descision', columns = ['venue', 'time'], values = 'target_achieved', aggfunc = 'count')
```

```python
table = df1.pivot_table(index=['Toss_descision'], columns = ['venue', 'time'], values = 'target_achieved', aggfunc = 'count')
table
```

```python
table = df1.pivot_table(index=['Toss_descision'], columns = ['venue', 'time'], values = 'target_achieved',\
                        aggfunc = 'count', fill_value = 0)
table
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
table.plot(kind='bar', figsize =(8,4))
plt.title("T20 ICC WC 2021", color = 'red')
plt.xlabel("Toss Decision", color = 'blue')
plt.ylabel("Number of Teams", color = 'green')
```

```python
table.plot(kind='barh', figsize =(6,4))
plt.title("T20 ICC WC 2021")
plt.xlabel("Number of Teams")
plt.ylabel("Toss Decision")
```
