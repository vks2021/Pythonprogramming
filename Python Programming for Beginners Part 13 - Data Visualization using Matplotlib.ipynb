```python
import matplotlib.pyplot as plt
plt.style.use('default')
import numpy as np
%matplotlib inline
# to create data for plotting
x_values = [0,1,2,3,4,5]
squares = [0,1,4,9,16,25]
cubes = [0,1,8,27,64,125]
#default graph style for plot is a line
plt.plot(x_values, squares, color="black", ms=5, marker="o", label="square")
plt.plot(x_values, cubes, color="red", ls="--", marker="o", ms=7, label="cube")
#plt.show
```

```python
import matplotlib.pyplot as plt
plt.style.use('default')
import numpy as np
%matplotlib inline
x = np.linspace(0, 5)
plt.plot(x,  .4 + x, color='blue')               
plt.plot(x, 2 + 2 * x, '--', color='#FFDD44')   
plt.title("Line Plot")
plt.xlabel("x-axis") 
plt.ylabel("y-axis")
```

```python
# to create data for plotting
x_values = [0,1,2,3,4,5]
squares = [0,1,4,9,16,25]
cubes = [0,1,8,27,64,125]
#default graph style for plot is a line
plt.plot(x_values, squares, color="black", ms=5, marker="o", label="square")
plt.plot(x_values, cubes, color="red", ls="--", marker="o", ms=7, label="cube")
plt.legend(loc = 'upper left', bbox_to_anchor=(1,1)) #to show the labels at proper location
plt.title("Line Plot")
plt.xlabel("x-axis") 
plt.ylabel("y-axis")
plt.xticks(list(range(6)), rotation="vertical")
```

```python
#create data for plotting
x_values = [6,5,8,2,3]
y_values = ["A", "B", "C", "D", "E"]
plt.bar(y_values,x_values, color = "red")
plt.show()
```

```python
#create data for plotting
x_values = [6,5,8,2,3]
y_val = ["A", "B", "C", "D", "E"]
# Adding an "h" after bar will flip the graph
plt.barh(y_val,x_values, color ="green")
plt.show()
```

```python
#create data for plotting
x_values = [0,1,2,3,4,5]
squares = [0,1,4,9,16,25]
cubes = [0,1,8,27,64,125]
plt.scatter(x_values,squares, s=30, color = "red")
plt.scatter(x_values,cubes, s=30, color = "green")
plt.show()
```

```python
#generate data and plot histogram 
x = [1,2,1,1,6,4,2,4,8,9,4,2,4,10,6,4,5,7,7,3,2,7,5,3,5,9,2,1]
num_bins = 10
#plot for a histogram
plt.hist(x, bins= 4, color='blue', alpha=0.1)
plt.show()
```

```python
#generate data and plot Pie chart 
years = [2016,2017,2018,2019,2020]
prod = [46.7, 22.2, 26.1, 27.9, 20.1]
fig1, ax1 = plt.subplots()
ax1.pie(prod, labels=years, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')

plt.show()
```
